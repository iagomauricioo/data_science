# Criei esse programa para tentar tratar pequenos dados com NumPy e Pandas

```python
import pandas as pd
import sqlite3
from datetime import datetime
```

```python
con = sqlite3.connect('users.db')
query = ('SELECT * from user;')
```

```python
df = pd.read_sql(query, con)
```

```python
df2 = df[["name", "email", "cpf", "phoneNumber", "birthday"]]
```
