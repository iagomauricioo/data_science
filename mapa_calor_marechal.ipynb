```python
import folium
from folium.plugins import HeatMap
import pandas as pd
from geopy.distance import geodesic
from IPython.display import HTML
import matplotlib.pyplot as plt
```

```python
arquivo_excel = '/content/drive/MyDrive/coordenadas.xlsx'
#arquivo_excel = '/content/coordenadas(1).xlsx'
df = pd.read_excel(arquivo_excel)
df = df[['tipo', 'cidade', 'coordenadas']]
df.dropna(inplace=True)
df
```

```python
data = df[['tipo', 'coordenadas']]
data[['latitude', 'longitude']] = data['coordenadas'].str.split(',', expand=True)
data['latitude'] = data['latitude'].astype(float)
data['longitude'] = data['longitude'].astype(float)
data = data.drop('coordenadas', axis=1)
data.dropna(inplace=True)
data
```

```python
barra_sao_miguel_coords = [-9.8379, -35.8979]
marechal_deodoro_coords = [-9.7092, -35.8964]
```

```python
center_lat = (barra_sao_miguel_coords[0] + marechal_deodoro_coords[0]) / 2
center_lon = (barra_sao_miguel_coords[1] + marechal_deodoro_coords[1]) / 2
center_coords = [center_lat, center_lon]
```

```python
m = folium.Map(location=[(data['latitude'].mean()), (data['longitude'].mean())], zoom_start=12)
```

```python
heat_data_mercado = [[row['latitude'], row['longitude']] for index, row in data.iterrows() if row['tipo'] == 'Mercado/Conven.']
heat_data_restaurante = [[row['latitude'], row['longitude']] for index, row in data.iterrows() if row['tipo'] == 'Restaurante/Bar']
heat_data_hotel = [[row['latitude'], row['longitude']] for index, row in data.iterrows() if row['tipo'] == 'Hotel/Pousada']
```

```python
HeatMap(heat_data_mercado, min_opacity=0.6, max_val=1.0, radius=25, blur=20, gradient={0.4: 'purple', 0.65: 'purple', 1: 'purple'}).add_to(m)
HeatMap(heat_data_restaurante, min_opacity=0.6, max_val=1.0, radius=25, blur=20, gradient={0.4: 'red', 0.65: 'red', 1: 'red'}).add_to(m)
HeatMap(heat_data_hotel, min_opacity=0.6, max_val=1.0, radius=25, blur=20, gradient={0.4: 'orange', 0.65: 'orange', 1: 'orange'}).add_to(m)
```

```python
legend_html = '''
     <div style="
     position: fixed;
     bottom: 50px; left: 50px; width: 200px; height: 150px;
     background-color: white; z-index:9999; font-size:14px;
     border:2px solid grey; padding: 10px;
     ">
     <h4>Legenda</h4>
     <div><span style="background-color:rgba(255,0,0,0.6); display:inline-block; width:20px; height:20px;"></span> Restaurante/Bar</div>
     <div><span style="background-color:rgba(128,0,128,0.6); display:inline-block; width:20px; height:20px;"></span> Mercado/Conven.</div>
     <div><span style="background-color:rgba(255,165,0,0.6); display:inline-block; width:20px; height:20px;"></span> Hotel/Pousada</div>
     </div>
     '''
m.get_root().html.add_child(folium.Element(legend_html))
```

```python
radius = 10000  # 10 km em metros
folium.Circle(location=center_coords, radius=radius, color='red', fill=True, fill_color='gray', fill_opacity=0.2).add_to(m)
```
