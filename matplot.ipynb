```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
x = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y = [0, 2, 3, 4, 2, 6, 7, 8, 12]
plt.plot(x, y)
plt.show()
```

```python
x = np.arange(-100, 100, 2)
y = x ** 2 - 2000
plt.plot(x, y)
plt.show()
```

```python
days = np.arange(1, 31, 1)
vacinados = np.random.randint(100, 1000, 30)
contagios = np.random.randint(100, 500, 30)
mortes = np.random.randint(10, 100)
plt.plot(days, vacinados)
plt.show()
print(days)
print(vacinados)
```

```python
dados = pd.DataFrame(
    {
        "days": days,
        "vacinados": vacinados,
        "contagios": contagios,
        "mortes": mortes,
    }
)

```

```python
dados
```

```python
dados.plot(kind="bar", x="days", y="vacinados")
plt.plot(days, contagios, "r")
plt.legend(["Contagios","Vacinados"])
plt.show()
```

```python
colors = ["red", "blue"]
sns.barplot(x="days", y="vacinados", data=dados, color=colors[1])
sns.barplot(x="days", y="contagios", data=dados, color=colors[0])
plt.legend(["Contagios","Vacinados"], labelcolor=colors)
plt.show()
```
