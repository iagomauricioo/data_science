```python
df_chuva = df[['Data', 'PRECIPITAȃO TOTAL, HORIO (mm)']]
df_chuva.rename(columns={
    'Data': 'Data',
    'PRECIPITAȃO TOTAL, HORIO (mm)': 'Chuva (mm)'
}, inplace=True)
df_chuva.Data = pd.to_datetime(df_chuva.Data)
df_chuva.set_index('Data', inplace=True)
df_chuva
```

```python
plt.figure(figsize=(20, 12))
plt.plot(df_chuva, color='green', marker='o')
plt.title('Série Histórica de Precipitação')
plt.xlabel('Período')
plt.ylabel('Precipitação (mm)')
plt.grid(True)
plt.show()
```


O gráfico exibido é uma série histórica de precipitação, mostrando a quantidade de chuva (em milímetros) ao longo do tempo (no eixo X). Cada ponto azul representa a precipitação em um período específico, com os valores variando principalmente entre 0 e 20 mm.
